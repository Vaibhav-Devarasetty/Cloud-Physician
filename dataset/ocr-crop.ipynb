```python
df = pd.read_csv('Nihon-Kohden-lifescope-A.csv')
df.head()
```

```python
def convertString2List(s):
    s = str(s)
    a = ''
    for x in s:
        if x not in'[]':
            a+=x
    return list(map(float, a.split(', ')))
```

```python
image_Path = '/home/sharchit/Desktop/ml/cloud-physician/Nihon-Kohden-lifescope-A/'
save_Path = '/home/sharchit/Desktop/ml/cloud-physician/NK-HR/'
iterations = 50
for i in range(iterations):
    imageName = df.iloc[i]['image_name']
    im = Image.open(image_Path+imageName)
    imageWidth, imageHeight = im.size
    x_center, y_center, width, height = convertString2List(df.iloc[i]['HR'])
    x_center *= imageWidth
    width *= imageWidth
    y_center *= imageHeight
    height *= imageHeight
    crop_im = im.crop((x_center - width/2, y_center - height / 2, x_center + width/2, y_center + height/2)).
    crop_im.save(save_Path + imageName)
```
